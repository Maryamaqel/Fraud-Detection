# 3- Preprocessing Data

```python
#drop unnecessary columns for both train and test 
train_data.drop("Unnamed: 0",axis=1,inplace=True) 
test_data.drop("Unnamed: 0",axis=1,inplace=True) 
train_data = train_data.drop(['cc_num','first','last','trans_num'],axis=1)
test_data = test_data.drop(['cc_num','first','last','trans_num'],axis=1)
```

```python
#adding trans_date column to extract only the date from trans_date_trans_time column in test data
train_data["trans_date_trans_time"] = pd.to_datetime(train_data["trans_date_trans_time"])
train_data["trans_date"] = train_data["trans_date_trans_time"].dt.date
train_data["trans_date"]= pd.to_datetime(train_data["trans_date"])
```

```python
#extracting month,year,day and adding a column for them from trans_date from train dataset
train_data['year'] = train_data['trans_date'].dt.year
train_data['month'] =train_data['trans_date'].dt.month
train_data['day'] = train_data['trans_date'].dt.day
```

```python
#adding trans_date column to extract only the date from trans_date_trans_time column in test data
test_data["trans_date_trans_time"] = pd.to_datetime(test_data["trans_date_trans_time"])
test_data["trans_date"] = test_data["trans_date_trans_time"].dt.date
test_data["trans_date"]= pd.to_datetime(test_data["trans_date"]) 
```

```python
#extracting month,year,day and adding a column for them from trans_date from test dataset
test_data['year'] = test_data['trans_date'].dt.year
test_data['month'] = test_data['trans_date'].dt.month
test_data['day'] = test_data['trans_date'].dt.day
```

```python
#encode categorical data
train_data = pd.get_dummies(train_data,columns=['category'],drop_first=True)
test_data = pd.get_dummies(test_data,columns=['category'],drop_first=True)
```

```python
#remove spaces 
train_data.columns = train_data.columns.str.replace(' ', '')
test_data.columns = test_data.columns.str.replace(' ', '')
```

```python
#we make data set from only number data types to put in model
train = train_data.select_dtypes(include='number')
test = test_data.select_dtypes(include='number')
```

```python
#concatanation operation for x and y
total = pd.concat([train, test])
```

```python
#X for features and y for output label
X = total.drop("is_fraud",axis=1) 
y = total["is_fraud"]
```

```python
#drop columns that are irrelevant
X = X.drop(['zip','lat','long','unix_time','merch_lat','merch_long'],axis=1)
```

# 4-Building the Classification Model

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(
 X, y, test_size=0.3, random_state=42)
```

```python
#scale data into range
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train1 = scaler.fit_transform(X_train1)
X_test1 = scaler.transform(X_test1)
```

Decision Tree model

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train1,y_train1)
```

Model evaluation

```python
print("Score the X-train with Y-train is : ", dtc.score(X_train1,y_train1))
print("Score the X-test  with Y-test  is : ", dtc.score(X_test1,y_test1))

y_pred=dtc.predict(X_test1)
```

```python
print("Accuracy score " , accuracy_score(y_test1,y_pred))
print("F1 score: ", round(f1_score(y_test1, y_pred, average='weighted')*100,2),"%")
```

```python
matrix = classification_report(y_test1,y_pred,labels=[1,0])
print('Classification report : \n',matrix)
```

```python
print('Precision: %.3f' % precision_score(y_test1, y_pred))
```

```python
print("F1 score: ", round(f1_score(y_test1, y_pred, average='weighted')*100,2),"%")
```
